```python
import pandas as pd
from pathlib import Path

def get_placer_data(tag, dir):
    base = dir
    model_path = f'{base}/{tag}_model.pdb'
    df = pd.read_csv(f'{base}/{tag}.csv')
    df_good = df[df['plddt'] > 0.8]
    df_mean = df_good[['fape', 'lddt', 'rmsd', 'kabsch', 'prmsd', 'plddt', 'plddt_pde']].mean()
    mean_values = df_mean.to_dict()

    placer_item = {
        'path': model_path,
        'fape': mean_values['fape'],
        'lddt': mean_values['lddt'],
        'rmsd': mean_values['rmsd'],
        'kabsch': mean_values['kabsch'],
        'prmsd': mean_values['prmsd'],
        'plddt': mean_values['plddt'],
        'plddt_pde': mean_values['plddt_pde'],
        'good_count': len(df_good),
    }
    return placer_item

df = pd.read_parquet('rmsd_3rd.parquet')
def process(df, placer_dir, name='placer', col=('lmpnn', 'tag')):
    df['placer_data'] = df[col].apply(lambda x: get_placer_data(x, placer_dir))
    df_exp = pd.json_normalize(df['placer_data'])
    df_exp.columns = pd.MultiIndex.from_product([[name], df_exp.columns])
    df = pd.concat([df.drop(columns=['placer_data']), df_exp], axis=1)
    df.columns = pd.MultiIndex.from_tuples([
        col if isinstance(col, tuple) else (col, '')
        for col in df.columns
    ])
    return df

placer_dir = '../7_placer/output'
df = process(df, placer_dir, name='placer', col=('lmpnn', 'tag'))
df
```
